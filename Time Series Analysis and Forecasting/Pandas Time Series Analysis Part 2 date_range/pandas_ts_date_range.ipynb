<h1 style="color:blue;" align="center">Pandas Time Series Analysis Tutorial: date_range</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv("aapl_no_dates.csv")
df.head()
```

```python
df.shape
```

```python
rng = pd.date_range(start="6/1/2017",end="6/30/2017",freq='B')
rng
```

```python
rng.shape
```

```python
df.set_index(rng, inplace=True)
df.head()
```

```python
df.columns
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df.loc[(df.index >= "2017-06-01") & (df.index <="2017-06-10")]
```

```python
df.loc[(df.index >= "2017-06-01") & (df.index <="2017-06-10")].Close.mean()
```
